```python
def calcula_par(valor, juros, desc, par_min, par_max):
  
    valor_max = valor+(valor*juros/100)
    valor_min = valor-(valor*desc/100)
    
    num_min_par = round((valor / par_max) + 0.5)
    num_max_par = round((valor / par_min) - 0.5)

    if num_min_par == 1:
        num_max_par = round((valor_min / par_min) - 0.5)

    if num_max_par == 0:
        num_max_par = num_min_par

    if valor_min < num_min_par*par_max:
        num_min_par = round((valor_min / par_max) + 0.5)

    if valor_max > num_max_par*par_min and num_min_par != num_max_par:
        num_max_par = round((valor_max / par_min) - 0.5)

    return num_min_par, num_max_par
```
